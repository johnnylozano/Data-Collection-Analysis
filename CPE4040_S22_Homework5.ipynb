<h4>4. Solve a simiar problem as in Homework #3: Use GroupBy to find the distribution of rating in each genre.</h4>
<h4>   a. How many apps in 'Games' are for age 17 and above?</h4>
<h4>   b. How many apps in 'Education' are for kids 9 years and older?</h4>

```python
# Games
age = df_apps.groupby('cont_rating')
adult = age.get_group('17+')

adult_genre = adult.groupby('prime_genre')
adult_games = adult_genre.get_group('Games')
print('The number of Games for 17+ are: ',adult_games.size)


# Education
age = df_apps.groupby('cont_rating')
kids = age.get_group('9+')

kids_genre = kids.groupby('prime_genre')
kids_edu = kids_genre.get_group('Education')
print('The number of Education apps for 9+ are: ',kids_edu.size)
```

<h3>Q4: Practicing GroupBy and Data Aggregation (30 Points)</h3>

<h4>1. Import the dataset from the URL. This is a set of statistics of age, gender and occupation.</h4>
<h4> How many different occupations are in the dataset?

```python
users = pd.read_table('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', 
                      sep='|', index_col='user_id')
```

```python
occu = users['occupation'].value_counts()
print('Number of different occupations: ',occu.size)
print(occu)
```

<h4>2. Calculate the average age for each occupation.</h4> 

```python
occupation = users['occupation'].unique()

mean_occupation = {}
for a in occupation:
    occupation_group = users[users['occupation'] == a]
    mean = occupation_group['age'].mean()
    mean_occupation[a] = mean
print('The average age for each occupation is: ')    
mean_occupation
```

<h4>3. Calculate the male ratio for each occupation and sort it in descending order.</h4>

```python
# Number of Males per Occupation
gender = users.groupby('gender')
male = gender.get_group('M')

male_occupation = male.groupby('occupation')



# Number of Women per Occupation
gender = users.groupby('gender')
fem = gender.get_group('F')

fem_occupation = fem.groupby('occupation')


# Total calculation
total = (male_occupation.size()/(male_occupation.size()+fem_occupation.size()) )*100

male_ratio = total.sort_values()
print('The male ratio for each occupation is:')
male_ratio.fillna(100)
```

<h4>4.  For each occupation, calculate the minimum and maximum ages.</h4>

```python
occupation = users.groupby('occupation')
age = occupation['age']
print('The max age for each occupation is:')
print(age.max())
print()
print('The min age for each occupation is:')
print(age.min())
```

<h4>5. For each combination of occupation and gender, calculate the mean age.</h4>

```python
male_age = male_occupation['age']
print('The mean age for men for each occupation is:')
print(male_age.mean())

print()
fem_age = fem_occupation['age']
print('The mean age for women for each occupation is:')
print(fem_age.mean())
```
