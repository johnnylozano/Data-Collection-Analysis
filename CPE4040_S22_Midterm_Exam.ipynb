<h3> Q1: Happy Twosday! (13 points) </h3>

Yesterday, 2/22/22, is a rare day that all five digits of the date, 22222, are twos. The string is both palindrome and symmetric. Recall that 
   - A sequence is said to be palindrome if one half of the string is the reverse of the other half, e.g., "madam".
   - A sequence is said to be symmetrical if both halves of the string are the same, e.g., "abcabc"

Q1.1 Write a Python program to confirm that yesterday's date (22222) is indeed palindrome and symmetrical. 

Q1.2 There are more dates in February this year that are also palindrome and symmetric (such as 22122). Write a Python loop to find out all the other dates.

```python
# Answer for Q1-1

def is_palin_symm(string):
    half = int(len(string)/2)
 
    if len(string) % 2 == 0:  # even
        first_half = string[:half]
        second_half = string[half:]
    else:  # odd
        first_half = string[:half]
        second_half = string[half+1:]
# symmetric
    if first_half == second_half:
        print(string, 'is symmertical')
    else:
        print(string, 'is not symmertical') 
# palindrome
    if first_half == second_half[::-1]: 
        print(string, 'is palindrome')
    else:
        print(string, 'is not palindrome')


is_palin_symm('22222')
```

```python
# Answer for Q1-2

# There are 28 days on Feb, 2022, and the format of the dates are 2xx22, where xx is from 1 to 28

for day in range(1,29):
    date = 20022 + day*100
    is_palin_symm(str(date))
```

<h3>Q2: Generalized Checkerboard (12 points)</h3>

```python
# Using SLicing Method

def checkerboard(m,n):
    arr = np.zeros((m,n),dtype=int)
    arr[::2,::2] = 1    # [1,0,1,0..] on even-numbered rows
    arr[1::2,1::2] = 1  # [0,1,0,1..] on odd-numbered rows
    return arr

checkerboard(9, 12)
```

<h3>Q3: Sine Wave and Square Wave (20 points)</h3>

A continuous sine wave is given by y(t) = A*sin(2*pi*f*t), where A=amplitude and f=center frequency.
In order to digitize the continuous signal, we sample the signal with sampling rate fs.

Q3.1 Using the generate_sine_wave() function below to generate a 5-second sine wave with f=20Hz and sampling rate fs=100Hz. Use Matplotlib to plot the sine wave.

```python
def generate_sine_wave(freq, sample_rate, duration):
    t = np.linspace(0, duration, sample_rate * duration, endpoint=False)
    y = np.sin((2 * np.pi) * freq * t)
    return t, y
```

```python
# Answer for Q3.1

fs = 100
duration = 5   # 5-second 
f = 20    # the center frequency
time, y = generate_sine_wave(f, fs, duration)

plt.figure(figsize=(12,6))
plt.plot(time, y)
plt.xlabel('Second')
plt.ylabel('Amplitude')
plt.show()
```

```python
# Hint: Use a boolean array to check if the signal in the sine wave array is greater or less than 0

# use booean array
# sq = y.copy()  
# sq[sq > 0] = 1
# sq[sq < 0] = -1

sq = y.copy()      # This is not necessary. I just want to keep a copy of the original array
print (type(sq))

sq[sq >=0] = 1
sq[sq <0] = -1

plt.figure(figsize=(12,6))
plt.plot(time, sq, 'r')
plt.title("Square wave")
plt.grid(True, which='both')

plt.show()
```

```python
# Overlaying the two signals together

plt.figure(figsize=(12,6))
plt.plot(time, y, time, sq, 'r')
plt.xlim(0, 0.5)
plt.title("Sine Wave vs. Square wave")
plt.grid(True, which='both')

plt.show()
```
