```python
import numpy as np

# Plot three sets of data
# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# cyan dash+, blue squares and green triangles
plt.plot(t, t, 'c-+', t, t**2, 'ms', t, t**3, 'g^')

plt.title('Plt Change Format Demo')
plt.xlabel('Time (Sec)')
plt.ylabel('y = f(t)')
plt.show()
```

```python
import matplotlib.pylot as plt

# Plot three sets of data
# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# cyan dash+, blue squares and green triangles
plt.figure(figsize=(12,6))

plt.plot(t, t, 'c-+', label='y = t')
plt.plot(t, t**2, 'ms', label='y = t^2')
plt.plot(t, t**3, 'g^', label='y = t^3')

plt.legend(loc = 'center left')

plt.title('Plt Change Format Demo')

plt.xlabel('Time (Sec)')
plt.ylabel('y = f(t)')

plt.xticks(np.arange(0, 6, step=0.5))

plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)

plt.show()
```

```python
# Subplot examples: vertically stacked 2 plots
x1 = np.linspace(0.0, 5.0, 50)
x2 = np.linspace(0.0, 2.0, 40)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

# Create a subplot grid with 2 rows and 1 column. 
# This is sunbplot 1
plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'o-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

# This is subplot 2
plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('Time (t)')
plt.ylabel('Undamped')

plt.show()
```

```python
# Subplot: 2 by 2 

# Some example data to display
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

plt.subplot(2,2,1)
plt.plot(x, y)
plt.title('Axis [0, 0]')

plt.subplot(2,2,2)
plt.plot(x, y, 'orange')
plt.title('Axis [0, 1]')

plt.subplot(2,2,3)
plt.plot(x, -y, 'g')
plt.title('Axis [1, 0]')

plt.subplot(2,2,4)
plt.plot(x, -y, 'r')
plt.title('Axis [1, 1]')

plt.show()
```

```python
# Stacked Bar Chart

labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 35, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

fig, ax = plt.subplots()

plt.bar(labels, men_means, label='Men')
plt.bar(labels, women_means, bottom=men_means, label='Women')

plt.ylabel('Scores')
plt.title('Scores by group and gender', size=16) # text size 16
ax.legend()

plt.show()
```

```python
# Stack bar charts: 4 data sets

x = ['A', 'B', 'C', 'D']
y1 = np.array([10, 20, 10, 30])
y2 = np.array([20, 25, 15, 25])
y3 = np.array([12, 15, 19, 6])
y4 = np.array([10, 29, 13, 19])
  
# plot 4 bars in stack manner
plt.bar(x, y1, color='r')     # bottom bar 
plt.bar(x, y2, bottom=y1, color='b')   # 2nd bar
plt.bar(x, y3, bottom=y1+y2, color='y') # 3rd bar 
plt.bar(x, y4, bottom=y1+y2+y3, color='g') # top bar

plt.xlabel("Teams", size =14)
plt.ylabel("Score", size =14)
plt.legend(["Round 1", "Round 2", "Round 3", "Round 4"])
plt.title("Scores by Teams in 4 Rounds", size=16)

plt.show()
```

```python
# Stacked bar chart using pandas Dataframe plot 
import pandas as pd
  
df = pd.DataFrame([['A', 10, 20, 10, 26], ['B', 20, 25, 15, 21], 
                   ['C', 12, 15, 19, 6], ['D', 10, 18, 11, 19]],
        columns=['Team', 'Round 1', 'Round 2', 'Round 3', 'Round 4'])

print(df)
  
# plot data in stack manner of bar type
df.plot(x='Team', kind='bar', stacked=True,
        title='Stacked Bar Graph by Dataframe')

plt.xticks(rotation=0)

plt.show()
```

```python
# Grouped bar chart using pandas Dataframe plot 
import pandas as pd
  
df = pd.DataFrame([['A', 10, 20, 10, 26], ['B', 20, 25, 15, 21], 
                   ['C', 12, 15, 19, 6], ['D', 10, 18, 11, 19]],
        columns=['Team', 'Round 1', 'Round 2', 'Round 3', 'Round 4'])

print(df)
  
# plot data in stack manner of bar type
df.plot(x='Team', kind='bar', stacked=False,
        title='Stacked Bar Graph by Dataframe')

plt.xticks(rotation=0)

plt.show()
```

```python
# Group bar chart: side-by-side bars along X-axis (or Y-axis)

labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 35, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

x_loc = np.arange(len(labels))  # the x locations for the groups
width = 0.4  # the total width of the bars 
  
plt.bar(x_loc - width/2, men_means, width, label = 'Men')
plt.bar(x_loc + width/2, women_means, width, label = 'Women')

# plt.xticks(): mapping tick location with the tick labels on x-Axis
plt.xticks(x_loc, labels)

plt.xlabel('Groups')
plt.ylabel('Scores')
plt.title('Scores by group and gender', size=16) 
plt.legend()

plt.show()
```

```python
# An example for histogram with normal random 

x = np.random.normal(0, 1, 10000)

# the histogram of the data: 100 bins (equally spaced)
plt.hist(x, bins=100, density=1, color='g')

plt.title('Normal Distribution N(0,1)')
plt.ylabel('Probability')
plt.xlabel('Values')

# Adding some text info on the plot. Size is the text size
plt.text(1.8, 0.35, 'mean=0, var=1', size=12)

# Specify the axes values = [xmin, xmax, ymin, ymax]
plt.axis([-5, 5, 0, 0.5])

plt.grid(True)
plt.show()
```

```python
# Pie chart, where the slices will be ordered and plotted 
#  counter-clockwise:

labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.2, 0, 0)  # only "explode" the 2nd slice by 0.2 (i.e. 'Hogs')

plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x,y): return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)

plt.contourf(X, Y, f(X,Y), 8, alpha=.75, cmap='jet')
C = plt.contour(X, Y, f(X,Y), 8, colors='black', linewidth=.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='hot')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.cm.hot)
ax.contourf(X, Y, Z, zdir='z', offset=-2, cmap=plt.cm.hot)
ax.set_zlim(-2,2)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x,y):
    return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)

plt.axes([0.025,0.025,0.95,0.95])

plt.contourf(X, Y, f(X,Y), 8, alpha=.75, cmap=plt.cm.hot)
C = plt.contour(X, Y, f(X,Y), 8, colors='black', linewidth=.5)
plt.clabel(C, inline=1, fontsize=10)

plt.xticks([]), plt.yticks([])

plt.show()
```
